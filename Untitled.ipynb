```python
X_encoded= pd.get_dummies(X, columns=["City",
                                      "Gender",
                                      "Senior_Citizen",
                                      "Partner",
                                      "Dependents",
                                      "Phone_Service",
                                      "Multiple_Lines",
                                      "Internet_Service",
                                      "Online_Security",
                                      "Online_Backup",
                                      "Device_Protection",
                                      "Tech_Support",
                                      "Streaming_TV",
                                      "Streaming_Movies",
                                      "Contract",
                                      "Paperless_Billing",
                                      "Payment_Method",
                                     ])
X_encoded.head()
```

```python
y.unique()
```

```python
sum(y)/len(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, random_state=42, stratify=y)
```

```python
sum(y_train)/len(y_train)
```

```python
sum(y_test)/len(y_test)
```

```python
clsf_xgb=xgb.XGBClassifier(missing=None, seed=42, use_label_encoder=False)
clsf_xgb.fit(X_train,
             y_train,
             verbose=True,
             early_stopping_rounds=10,
             eval_metric='aucpr',
             eval_set=[(X_test,y_test)]
            ) 
```

```python
plot_confusion_matrix(clsf_xgb,
                      X_test,
                      y_test,
                      values_format='d',
                      display_labels=["Did not leave","Left"]
                     )
```

```python
#ROUND 1
max_depth = [3, 4, 5]
learning_rate=[0.1,0.05,0.01]
gamma=[0,0.25,1]
reg_lamda=[0,1.0,10.0]
scale_pos_weight=[1,3,5]
#ROUND 2
max_depth = [4]
learning_rate=[0.1,0.5,1]
gamma=[0]
reg_lamda=[0]
scale_pos_weight=[3]

param_grid = {
    'max_depth': max_depth,
    'learning_rate':learning_rate,
    'gamma':gamma,
    'reg_lamda':reg_lamda,
    'scale_pos_weight':scale_pos_weight
    }

optimal_params= GridSearchCV(
                estimator=xgb.XGBClassifier(missing=None, 
                                            seed=42, 
                                            use_label_encoder=False, 
                                            subsample=0.9, 
                                            colsample_bytree=0.5),
                param_grid=param_grid,
                scoring='roc_auc',
                verbose=0,
                n_jobs=4,
                cv=3
                )

optimal_params.fit(X_train,
                 y_train,
                 early_stopping_rounds=10,
                 eval_metric='aucpr',
                 eval_set=[(X_test,y_test)],
                 verbose=False
            )
print(optimal_params.best_params_)
```

```python
clsf_xgb=xgb.XGBClassifier(missing=None, 
                           seed=42, 
                           use_label_encoder=False,
                           gamma=0,
                           learning_rate=0.1,
                           max_depth=4,
                           reg_lamda=0,
                           scale_pos_weight=3,
                           subsample=0.9, 
                            colsample_bytree=0.5
                          )
clsf_xgb.fit(X_train,
             y_train,
             verbose=True,
             early_stopping_rounds=10,
             eval_metric='aucpr',
             eval_set=[(X_test,y_test)]
            )
```

```python
plot_confusion_matrix(clsf_xgb,
                      X_test,
                      y_test,
                      values_format='d',
                      display_labels=["Did not leave","Left"]
                     )
```

```python
clsf_xgb=xgb.XGBClassifier(missing=None, 
                           seed=42, 
                           use_label_encoder=False,
                           gamma=0,
                           learning_rate=0.1,
                           max_depth=4,
                           reg_lamda=0,
                           scale_pos_weight=3,
                           subsample=0.9, 
                            colsample_bytree=0.5,
                           n_estimator=1,
                          )
clsf_xgb.fit(X_train,
             y_train
            )

bst=clsf_xgb.get_booster()

for importance_type in ('weight', 'gain', 'cover', 'total_gain', 'total_cover'):
    print('%s: ' % importance_type, bst.get_score(importance_type=importance_type))
    
node_params = {'shape': 'box',
               'style': 'filled, rounded',
               'fillcolor': '#78cbe',
              }
leaf_params = {'shape': 'box',
               'style': 'filled',
               'fillcolor': '#e48038',
              }
xgb.to_graphviz(clsf_xgb, num_trees=0, size="10,10",
               condition_node_params=node_params,
               leaf_node_params=leaf_params)
```
